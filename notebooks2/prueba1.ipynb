Armo columnas nuevas


```python
#armar la columna dia. (que dia de la semana es )
df_filtrado["fecha_destino_recorrido"] = pd.to_datetime(df_filtrado["fecha_destino_recorrido"])
df_filtrado["dia"] = df_filtrado["fecha_destino_recorrido"].dt.day_name()

#armar columna estacion del año (otoño, invierno, primavera, verano)

def get_season(date):
    month = date.month
    if month in [12, 1, 2]:
        return "Verano"
    elif month in [3, 4, 5]:
        return "Otoño"
    elif month in [6, 7, 8]:
        return "Invierno"
    else:
        return "Primavera"

df_filtrado["estacion"] = df_filtrado["fecha_destino_recorrido"].apply(get_season)

df_filtrado.head(50)
```

Hacer ventanas

```python
def get_time_slot(dt):
    hour = dt.hour
    minute = dt.minute
    if 0 <= minute < 30:
        start = f"{hour:02d}:00"
        end = f"{hour:02d}:30"
    else:
        start = f"{hour:02d}:30"
        end = f"{(hour + 1)%24:02d}:00"
    return f"{start} - {end}"

# Asegúrate de que la columna sea tipo datetime
df_filtrado["fecha_destino_recorrido"] = pd.to_datetime(df_filtrado["fecha_destino_recorrido"])

df_filtrado["delta_t"] = df_filtrado["fecha_destino_recorrido"].apply(get_time_slot)

#cambiar la columna hora por mañana, tarde, noche
def get_time_of_day(dt):
    hour = dt.hour
    if 6 <= hour < 12:
        return "Mañana"
    elif 12 <= hour < 18:
        return "Tarde"
    else:
        return "Noche"
    
df_filtrado["hora"] = df_filtrado["fecha_destino_recorrido"].apply(get_time_of_day)

#agregar columna es_hora_pico
def is_peak_hour(dt):
    hour = dt.hour
    if (7 <= hour < 10) or (17 <= hour < 20):
        return True
    else:
        return False



df_filtrado["es_hora_pico"] = df_filtrado["fecha_destino_recorrido"].apply(is_peak_hour)

print(df_filtrado.shape)
df_filtrado.head(50)

```

```python
#quiero contar cuantas llegadas hay por estacion en cada delta_t 
df_count = df_filtrado.groupby(["id_estacion_destino", "delta_t"]).size().reset_index(name='count')
#agregar las demas columnas que estaban en df_filtrado que son fecha_destino_recorrido	id_estacion_destino	delta_t	hora	es_hora_pico


#agregar columna fecha
df_count["fecha_destino_recorrido"] = df_filtrado["fecha_destino_recorrido"].unique()[0]
df_count["hora"] = df_filtrado["hora"].unique()[0]  
df_count["es_hora_pico"] = df_filtrado["es_hora_pico"].unique()[0]
df_count["dia"] = df_filtrado["dia"].unique()[0]
df_count["estacion"] = df_filtrado["estacion"].unique()[0]




print(df_count.head(50))
print(df_count.shape)
```
