```python
import pandas as pd

df = pd.read_csv("../data/processed/recorridos/trips_2023.csv")

print("estaciones totales:", df["id_estacion_destino"].unique())
# Contar la cantidad de viajes por estación destino
estaciones_count = df["id_estacion_destino"].value_counts()

# Ordenar estaciones por cantidad de viajes (de mayor a menor)
estaciones_count = estaciones_count.sort_values(ascending=False)

# Calcular el 80% del total de viajes
total_viajes = estaciones_count.sum()
umbral_viajes = total_viajes * 0.80

# Seleccionar estaciones que suman hasta el 80% de los viajes
viajes_acumulados = estaciones_count.cumsum()
estaciones_preservadas = estaciones_count[viajes_acumulados <= umbral_viajes].index

# Si el último valor no llega justo al 80%, agregamos la siguiente estación para superar el umbral
if viajes_acumulados[estaciones_preservadas[-1]] < umbral_viajes:
    estaciones_preservadas = estaciones_count[viajes_acumulados <= umbral_viajes].index.tolist()
    estaciones_preservadas.append(estaciones_count.index[len(estaciones_preservadas)])

# Filtrar el DataFrame
df_filtrado = df[df["id_estacion_destino"].isin(estaciones_preservadas)]

# Printear los id de las estaciones preservadas
print("Estaciones preservadas:", df_filtrado["id_estacion_destino"].unique())

print(len(df_filtrado["id_estacion_destino"].unique()))
# Ordenar por fecha_destino_recorrido
df_filtrado = df_filtrado.sort_values(by=["fecha_destino_recorrido"], ascending=True)

# Quedarse solo con las columnas necesarias
df_filtrado = df_filtrado[["fecha_destino_recorrido", "id_estacion_destino"]]

print(df_filtrado.head(50))

```

Armo columnas nuevas


```python
#armar la columna dia. (que dia de la semana es )
df_filtrado["fecha_destino_recorrido"] = pd.to_datetime(df_filtrado["fecha_destino_recorrido"])
df_filtrado["dia"] = df_filtrado["fecha_destino_recorrido"].dt.day_name()

#armar columna estacion del año (otoño, invierno, primavera, verano)

def get_season(date):
    month = date.month
    if month in [12, 1, 2]:
        return "Verano"
    elif month in [3, 4, 5]:
        return "Otoño"
    elif month in [6, 7, 8]:
        return "Invierno"
    else:
        return "Primavera"

df_filtrado["estacion"] = df_filtrado["fecha_destino_recorrido"].apply(get_season)

df_filtrado.head(50)
```

Hacer ventanas

```python
def get_time_slot(dt):
    hour = dt.hour
    minute = dt.minute
    if 0 <= minute < 30:
        start = f"{hour:02d}:00"
        end = f"{hour:02d}:30"
    else:
        start = f"{hour:02d}:30"
        end = f"{(hour + 1)%24:02d}:00"
    return f"{start} - {end}"

# Asegúrate de que la columna sea tipo datetime
df_filtrado["fecha_destino_recorrido"] = pd.to_datetime(df_filtrado["fecha_destino_recorrido"])

df_filtrado["delta_t"] = df_filtrado["fecha_destino_recorrido"].apply(get_time_slot)

#cambiar la columna hora por mañana, tarde, noche
def get_time_of_day(dt):
    hour = dt.hour
    if 6 <= hour < 12:
        return "Mañana"
    elif 12 <= hour < 18:
        return "Tarde"
    else:
        return "Noche"
    
df_filtrado["hora"] = df_filtrado["fecha_destino_recorrido"].apply(get_time_of_day)

#agregar columna es_hora_pico
def is_peak_hour(dt):
    hour = dt.hour
    if (7 <= hour < 10) or (17 <= hour < 20):
        return True
    else:
        return False

df_filtrado["es_hora_pico"] = df_filtrado["fecha_destino_recorrido"].apply(is_peak_hour)


df_filtrado.head(50)
```
